```python
from transformers import VitsModel, AutoTokenizer
import torch
import numpy as np
import scipy.io.wavfile

name = "facebook/mms-tts-eng"

model = VitsModel.from_pretrained(name)
tokenizer = AutoTokenizer.from_pretrained(name)
```

```python

# Prepare the input textThe Norwegians are northerners, trundlers, southerners and people from all other regions. 
text  = " The Norwegians are northerners, trundlers, southerners and people from all other regions. The Norwegians are also immigrants from Afghanistan, Pakistan, Poland, Sweden, Somalia and Syria. The The Norwegians are girls who are fond of girls, boys who are fond of boys, and girls and boys who are fond of each other. The Norwegians believe in God, Allah, everything and nothing. The Norwegians like Grieg, Hugo, Helbylis and Kari Bremnes. In other words, Norway is you. Norway is us. My biggest hope for Norway is that we will be able to take care of each other. That we will continue to build this country on trust, community and generosity."

inputs = tokenizer(text, return_tensors="pt")
# Generate the waveform
with torch.no_grad():
    output = model(**inputs).waveform

# Convert the output to a numpy array
output_np = output.cpu().numpy()

# If the output is in the range [-1, 1], rescale it to the range of int16

# if output_np.dtype == np.float32:
#     output_np = (output_np * 32767).astype(np.int16)

# Write the audio data to a WAV file
scipy.io.wavfile.write("sound/king's_speech_1.wav", rate=model.config.sampling_rate, data=output_np[0])
```
